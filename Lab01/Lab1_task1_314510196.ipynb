```python
import time
import numpy as np

# === 超參數設定 ===
EPOCH = 20
Batch_size = 500
Learning_rate = 5e-3  # 固定學習率

# === 資料預處理 ===
train_data = train_data / 255.0

# 切出訓練集 / 驗證集
train_X = train_data[:train_image_num - val_image_num]
val_X   = train_data[train_image_num - val_image_num:]

train_Y = train_label[:train_image_num - val_image_num]
val_Y   = train_label[train_image_num - val_image_num:]

train_label_onehot = np.eye(10)[train_Y]
val_label_onehot   = np.eye(10)[val_Y]

train_batch_num = len(train_X) // Batch_size
val_batch_num   = len(val_X) // Batch_size

# === 建立模型 ===
net = model.Network()

# 驗證集歷史
val_loss_history = []
min_val_loss = np.inf
min_val_loss_epoch = 0
best_val_acc = 0

patience = 10 
wait = 0

# === Debug 訓練迴圈 ===
for epoch in range(1, EPOCH + 1):
    
    total_train_loss = 0
    train_hit = 0
    start_time = time.time()
    # print(f"\n=== Epoch {epoch} ===")
    
    for it in range(train_batch_num):
        batch_x = train_X[it*Batch_size:(it+1)*Batch_size]
        batch_y = train_label_onehot[it*Batch_size:(it+1)*Batch_size]

        # 前向傳播
        pred, train_loss = net.forward(batch_x, batch_y, training=True)

        # 計算 batch 正確率
        pred_index = np.argmax(pred, axis=1)
        train_hit += (pred_index == train_Y[it*Batch_size:(it+1)*Batch_size]).sum()
        total_train_loss += train_loss

        # Debug 梯度更新前後
        # before = net.layers[0].weights.copy()
        net.backward()
        net.update(Learning_rate)
        # after = net.layers[0].weights
        # print(f"[DEBUG] Iter {it+1}/{train_batch_num} | Batch Loss={train_loss:.4f} | Weight diff={np.sum(np.abs(after - before)):.6f}")

    # 計算整個 epoch 的訓練準確率與平均 loss
    train_acc = train_hit / len(train_X) * 100
    avg_train_loss = total_train_loss / train_batch_num

    
    # --- Validation (一次 forward 整個驗證集) ---
    val_pred, val_loss = net.forward(val_X, val_label_onehot, training=False)
    val_pred_index = np.argmax(val_pred, axis=1)
    val_acc = np.mean(val_pred_index == val_Y) * 100

    val_loss_history.append(val_loss)

    # 儲存最佳模型
    if val_loss < min_val_loss:
        print(f'Task1 | Minimum validation loss in history. Saving model parameters at Epoch {epoch}...')
        net.save_params()
        min_val_loss = val_loss
        min_val_loss_epoch = epoch
        best_val_acc = val_acc
        wait = 0 # reset patience 
    else: 
        wait += 1 
        if wait >= patience: 
            print(f"Early stopping triggered at Epoch {epoch}. Best epoch was {min_val_loss_epoch}.") 
            break

    end_time = time.time()
    print(f'Task1 | Epoch {epoch:03d} | Train Loss: {avg_train_loss:.4f} | Train Acc: {train_acc:.2f}% | '
          f'Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}% | Epoch time: {end_time - start_time:.2f} sec')

print(f'Task1 | Minimum validation loss: {min_val_loss:.4f} at Epoch {min_val_loss_epoch} | Val Acc: {best_val_acc:.2f}%')

    
```
